### Creación nueva columna: 'continent'

```python
import country_converter as coco

df.columns = df.columns.str.strip()

# Creación de la columna continente, rellenándola según el país:
cc = coco.CountryConverter()
df["Continent"] = df["Country"].apply(lambda x: cc.convert(x, to="continent"))
```

```python
# Creación del nuevo CSV
df.to_csv("Social Progress Index 2022.csv", index=False)
```
